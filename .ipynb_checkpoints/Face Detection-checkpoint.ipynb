```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
img = cv2.imread('data/testimage.jpg')
```

```python
img.shape
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray, cmap='gray')
```

```python
haar = cv2.CascadeClassifier('data/haarcascade_frontalface_default.xml')
```

```python
faces = haar.detectMultiScale(gray)
print(faces)
```

```python
cv2.rectangle(img, (221, 244), (221+480, 244+480), (0,255,0),3)
```

```python
plt.imshow(img)
```

```python
cv2.imshow('detection_frame', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
